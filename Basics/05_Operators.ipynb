## Order of Operations

Parentheses `()`: Expressions within parentheses are evaluated first.

Exponents `**`: Exponentiation operations are performed next.

Multiplication `*`, Division `/`, Floor Division `//`, Modulus `%`: These operators have the same precedence and are evaluated from left to right. 

Addition `+`, Subtraction `-`: These operators have the lowest precedence among arithmetic operators and are also evaluated from left to right.

```python
result = 8 / 4 + 5 * (3 + 1) - 7 % 4

print(result)
```

### Evaluation steps:

- Parentheses: (3 + 1) evaluates to 4.

**result = 8 / 4 + 5 * 4 - 7 % 4**

- Multiplication, Division, Modulus (left to right):
    - 8 / 4 evaluates to 2.0.
    - 5 * 4 evaluates to 20.
    - 7 % 4 evaluates to 3.
result = 2.0 + 20 - 3
- Addition and Subtraction (left to right):
    - 2.0 + 20 evaluates to 22.0.
    - 22.0 - 3 evaluates to 19.0.
### ***result = 19.0***

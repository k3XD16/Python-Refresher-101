## Conditional Statements

```python
age = int(input("Enter age: "))

if age < 18:
    print("You're too Young! not eligible for Google Gemini Pro ₹35000 worth")
elif age <= 25 :
    print("You're eligible for Google Gemini Pro ₹35000 worth")
else:
    print("You're too Old!! not eligible for Google Gemini Pro ₹35000 worth")
```

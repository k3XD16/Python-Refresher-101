```python
#same values
job_title = 'Data Analyst'
job_title2 = 'Data Analyst'
```

```python
print(id(job_title))
print(id(job_title2))  
```

```python
job_title3 = job_title
```

```python
print(id(job_title), id(job_title3))    
```


- `job_title2 = 'Data Analyst'` → creates a new string object

- `job_title3 = job_title` → points to the same string object

That’s why `job_title3` and `job_title` share the same `id()`, but `job_title2` doesn’t.

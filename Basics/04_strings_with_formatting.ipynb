```python
skill = 'Python'
```

```python
print(skill.upper())    #upper()
print(skill.lower())    #lower 
print(skill.capitalize())   #captialize
print(skill.isalnum())      #check is Alpha(r)Numeric
print(skill.startswith('P'))    #case sensitive 
print(skill.endswith('N'))    #case sensitive 
print('Malayalam'.count('al'))  #count
print('')
```

```python
print(str.upper(skill)) #equal to skill.upper()

print(str.endswith(skill,'n'))
```

```python
print(str.replace(skill,'P','J'))
```

```python
job_title = 'Doto Analyst'
print(f'Before:     {job_title}')
#write a code
job_title = str.replace(job_title,'Doto','Data')

print(f'After:      {job_title}')
```
